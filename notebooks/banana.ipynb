```python
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns
```

```python
path = "C:\\Users\\acer\\School Stuffs\\AI Application\\Finals\\data\\banana_quality_dataset.csv"
data = pd.read_csv(path)
print(data)
```

```python
n_a_region = data[(data['region'] == "Costa Rica") | (data['region'] == "Guatemala") ]
s_a_region = data[(data['region'] == "Ecuador") | (data['region'] == "Colombia") | (data['region'] == "Brazil")]
asia_region = data[(data['region'] == "Philippines") | (data['region'] == "India")]
print(asia_region)
filt = s_a_region[['length_cm','region']]
g = sns.pairplot(filt,hue="region", kind = "scatter",palette="coolwarm")

filt = asia_region[[ 'length_cm','region']]
g = sns.pairplot(filt,hue="region", kind = "scatter",palette="coolwarm")

filt = n_a_region[[ 'length_cm','region']]
g = sns.pairplot(filt,hue="region", kind = "scatter",palette="coolwarm")

filt = data[[ "weight_g",'length_cm','region']]
g = sns.pairplot(filt,hue="region", kind = "scatter",palette="coolwarm")

```
