```python
# List of best numerical columns for correlation
selected_columns = [

    'quality_score',
    'ripeness_index',
    'sugar_content_brix',
    'firmness_kgf',
    'length_cm',
    'weight_g',
    'tree_age_years',
    'altitude_m',
    'rainfall_mm',
    'soil_nitrogen_ppm'
]

# Correlation matrix
correlation_matrix = data[selected_columns].corr().round(2)

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, linewidths=1.5)
plt.title('Correlation of stuff')
plt.show()
```
